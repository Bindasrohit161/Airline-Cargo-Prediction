```python
#creating  betweenness centrality
routs = airline.groupby([AustralianPort,ForeignPort]).size().reset_index(name='count')
g = nx.from_pandas_edgelist(routs, source=AustralianPort, target=ForeignPort)
Btwnessv = pd.DataFrame(pd.DataFrame([nx.betweenness_centrality(g)]).unstack())
Btwnessv = Btwnessv.reset_index().rename(columns = { 'level_0': AustralianPort, 'level_1':'useless',
                                                  0:Btwness}).drop('useless',axis=1)
airline = Btwnessv.merge(airline,on=AustralianPort,how='inner')


#creating   closeness centrality
g = nx.from_pandas_edgelist(routs, source=AustralianPort, target=ForeignPort)
closenessv = pd.DataFrame(pd.DataFrame([nx.closeness_centrality(g)]).unstack())
closenessv = closenessv.reset_index().rename(columns = { 'level_0': AustralianPort, 'level_1':'useless',
                                                  0:closeness}).drop('useless',axis=1)
airline = closenessv.merge(airline,on=AustralianPort,how='inner')



#creating   degree centrality
g = nx.from_pandas_edgelist(routs, source=AustralianPort, target=ForeignPort)
Deg_centralityv = pd.DataFrame(pd.DataFrame([nx.degree_centrality(g)]).unstack())
Deg_centralityv = Deg_centralityv.reset_index().rename(columns = { 'level_0': AustralianPort, 'level_1':'useless',
                                                  0:Deg_centrality}).drop('useless',axis=1)
airline = Deg_centralityv.merge(airline,on=AustralianPort,how='inner')






airline[Freight_In_Out_prct] = (airline[Freight_In_tonnes] - airline[Freight_In_tonnes])/(1+airline[Freight_In_tonnes])
airline[Passengers_In_Out_prct] = (airline[Passengers_In] - airline[Passengers_Out])/(1+airline[Passengers_In])
# airline['Passengers_Total_Inside'] = (airline['Passengers_In'] - airline['Passengers_Out'])

airline[Mail_In_Out_prct] = (airline[Mail_In_tonnes] - airline[Mail_Out_tonnes])/(1+airline[Mail_Total_tonnes])



airline[W_Passengers_In] = airline[Passengers_In] *68/1000

airline[W_Passengers_Out] = airline[Passengers_Out] *68/1000
airline[W_Passengers_Total] = airline[Passengers_Total] *68/1000

# airline['W_Passengers_Total_Inside'] = airline['Passengers_Total_Inside'] *64/1000
airline[flights_total] = airline[Passengers_Total]/450
airline[Total_wt] = airline[flights_total] * (airline[Passengers_Total] *68/1000 + airline[Freight_Total_tonnes]+ airline[Mail_Total_tonnes])
airline[Avg_flight_wt] = (airline[Passengers_Total] *68/1000 + airline[Freight_Total_tonnes]+ airline[Mail_Total_tonnes])/(1+airline[flights_total])

# airline['flights'] = airline['Passengers_Total']/250
# airline['flightst_total'] = airline['Passengers_Total']/250
```

```python
# Dropping all total zero rows data
all_zero = airline[(airline[Freight_Total_tonnes]  == 0) & (airline[Passengers_Total]  == 0) & (airline[Mail_Total_tonnes]  == 0)].index
airline = airline.drop(all_zero)
```
