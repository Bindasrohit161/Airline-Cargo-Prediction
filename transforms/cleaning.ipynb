```python
k = airline_grp.reset_index(drop=True).copy()
# IQR
Q1 = np.percentile(k[Passengers_Total], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(k[Passengers_Total], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", k.shape)
 
# Upper bound
upper = np.where(k[Passengers_Total] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(k[Passengers_Total] <= (Q1-1.5*IQR))
''' Removing the Outliers '''
k.drop(upper[0], inplace = True)
k.drop(lower[0], inplace = True)
 
print("New Shape: ", k.shape)
```

```python
k = k.reset_index(drop=True)# IQR
Q1 = np.percentile(k[Mail_Total_tonnes], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(k[Mail_Total_tonnes], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", k.shape)
 
# Upper bound
upper = np.where(k[Mail_Total_tonnes] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(k[Mail_Total_tonnes] <= (Q1-1.5*IQR))
''' Removing the Outliers '''
k.drop(upper[0], inplace = True)
k.drop(lower[0], inplace = True)
 
print("New Shape: ", k.shape)
```
